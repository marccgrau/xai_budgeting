```python
import pandas as pd
```

```python
# Specify the file path
qualtrics = 'C:/Users/domin/Documents/GitHub/xai_budgeting/001_data/003_experiment/001_budget/118_qualtrics_results.csv'

# Load the .csv file into a DataFrame
qualtrics_df = pd.read_csv(qualtrics)

# Drop the observations without comment
qualtrics_df = qualtrics_df.dropna(subset=['Scenario1_Explainer', 'AI_Resubmission14', 'NoAI_Resubmission14'], how='all')

#Print overview
len(qualtrics_df)
```

```python
qualtrics_df.head()
```

Create textfile for initial submissions

```python
# Drop rows where all specified columns are NaN or empty
# Replace '' with NaN to also consider empty strings as NaN
qualtrics_df.replace('', pd.NA, inplace=True)
qualtrics_df.dropna(subset=['Scenario1_Explainer'], how='all', inplace=True)

# Select the specific columns you're interested in
subset_df = qualtrics_df[['Scenario1_Explainer']]

# Save these columns to a text file, using tab as a separator
subset_df.to_csv('C:/Users/domin/Documents/GitHub/xai_budgeting/001_data/003_experiment/002_justifications/initial_submission.txt', sep='\t', index=False)

# 'index=False' is used to prevent pandas from writing the DataFrame index into the file
```

Create text file for resubmissions

```python
# Drop rows where all specified columns are NaN or empty
# Replace '' with NaN to also consider empty strings as NaN
qualtrics_df.replace('', pd.NA, inplace=True)
qualtrics_df.dropna(subset=[ 'AI_Resubmission14'], how='all', inplace=True)

# Select the specific columns you're interested in
subset_df = qualtrics_df[['AI_Resubmission14']]

# Save these columns to a text file, using tab as a separator
subset_df.to_csv('C:/Users/domin/Documents/GitHub/xai_budgeting/001_data/003_experiment/002_justifications/ai_resubmissions.txt', sep='\t', index=False)

# 'index=False' is used to prevent pandas from writing the DataFrame index into the file
```
