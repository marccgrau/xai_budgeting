```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style('whitegrid')
```

```python
file_path = "../data/final/merged_double_digit.csv"
```

```python
df = pd.read_csv(file_path, index_col=None, header=0)
```

```python
df.head()
```

```python
aggregated = df.groupby(['Year', 'Acc-ID']).agg({'Realized': 'sum', 'Budget y': 'sum'}).reset_index()
```

```python
# Unique Acc-IDs
unique_acc_ids = aggregated['Acc-ID'].unique()

for acc_id in unique_acc_ids:
    acc_data = aggregated[aggregated['Acc-ID'] == acc_id]
    
    # Create a plot for each Acc-ID
    plt.figure(figsize=(8, 4))
    plt.plot(acc_data['Year'], acc_data['Realized'], label='Realized', marker='o', linestyle='-')
    plt.plot(acc_data['Year'], acc_data['Budget y'], label='Budget y', marker='x', linestyle='--')
    
    plt.xlabel('Year')
    plt.ylabel('Value')
    plt.title(f'Realized and Budget y Over Years for Acc-ID {acc_id}')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    
    plt.show()
```
