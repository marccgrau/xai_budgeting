```python
import requests
import pandas as pd
```

```python
def get_budget_data_v2():
    # API endpoint URL
    url = 'https://daten.stadt.sg.ch/api/records/1.0/search/?dataset=budget-der-stadt-stgallen-nach-funktionen'

    # Define parameters for the API request
    params = {
        'limit': 20  # Number of records to fetch
    }

    # Make the GET request
    response = requests.get(url, params=params)

    # Check if the request was successful
    if response.status_code == 200:
        return response.json()  # Return the JSON response
    else:
        return f'Error: {response.status_code}'

# Call the function to get data
data = get_budget_data_v2()
```

```python
# Parse and create a dataframe
if isinstance(data, dict) and 'records' in data:
    # Extract the fields from each record
    records = data['records']
    df_data = []
    for record in records:
        fields = record.get('fields', {})
        df_data.append({
            'jahr': fields.get('jahr'),
            'konto_3_stellig_nr': fields.get('konto_3_stellig_nr'),
            'konto_3_stellig': fields.get('konto_3_stellig'),
            'direktion_nr': fields.get('direktion_nr'),
            'direktion': fields.get('direktion'),
            'dienststelle_nr': fields.get('dienststelle_nr'),
            'dienststelle': fields.get('dienststelle'),
            'bereich_nr': fields.get('bereich_nr'),
            'bereich': fields.get('bereich'),
            'funktion_nr': fields.get('funktion_nr'),
            'bezeichnung': fields.get('bezeichnung'),
            'ergebnis': fields.get('ergebnis')
        })

    # Create the dataframe
    df = pd.DataFrame(df_data)

    # Display the dataframe
    print(df)
else:
    print("Error or unexpected format in API response:", data)
```
