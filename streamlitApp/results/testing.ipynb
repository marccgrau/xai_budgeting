```python
import pandas as pd
```

```python
df = pd.read_csv('lstm_results.csv')

#keep column Year, Region, Acc-ID, y, Budget y and yhat
df = df[['Year', 'Region', 'Acc-ID', 'y', 'Budget y', 'yhat']]
#rename yhat to yhat_lstm
df = df.rename(columns={'yhat': 'yhat_lstm'})
df
```

```python
dfcat = pd.read_csv('catboost_results.csv')
#merge df and dfcat on y = Realized in dfcat and Budget y = Budget y

df = pd.merge(df, dfcat, how='inner', left_on=['y', 'Budget y'], right_on=['Realized', 'Budget y'])
#rename prediction column to yhat_cat
df = df.rename(columns={'Predicted': 'yhat_cat'})
#drop Realized column
df = df.drop(columns='Realized')
df
```

```python
dfXg = pd.read_csv('xgboost_results.csv')
#merge df and dfxg on y = Realized in dfxg and Budget y = Budget y

df = pd.merge(df, dfXg, how='inner', left_on=['y', 'Budget y'], right_on=['Realized', 'Budget y'])
#rename prediction column to yhat_xg
df = df.rename(columns={'Predicted': 'yhat_xg'})
#drop Realized column
df = df.drop(columns='Realized')
df
```

```python
dfRandom = pd.read_csv('randomforest_results.csv')
#merge df and dfRandom on y = Realized in dfRandom and Budget y = Budget y

df = pd.merge(df, dfRandom, how='inner', left_on=['y', 'Budget y'], right_on=['Realized', 'Budget y'])
#rename prediction column to yhat_rf
df = df.rename(columns={'Predicted': 'yhat_rf'})
#drop Realized column
df = df.drop(columns='Realized')
df
```

```python
dfProph = pd.read_csv('prophet_results.csv')
#keep columns Region, Acc-ID, yhat
dfProph = dfProph[['Region', 'Acc-ID', 'yhat']]
#rename yhat to yhat_proph
dfProph = dfProph.rename(columns={'yhat': 'yhat_proph'})

#merge df and dfProph on Region and Acc-ID
df = pd.merge(df, dfProph, how='inner', on=['Region', 'Acc-ID'])
#drop duplicates
df = df.drop_duplicates()
#keep only unique combinations of Region and Acc-ID
df = df.drop_duplicates(subset=['Region', 'Acc-ID'])
df
```
